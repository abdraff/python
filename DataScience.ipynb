import modules pandas dan numpy

```
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

Membuka dan membaca data dari excel

```
df_can = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)

print ('Data read into a pandas dataframe!')
```

Melihat 5 baris dari dataset menggunakan fungsi head()
NOTE: Jika ingin melihat 10 baris, gunakan head(10)

```
df_can.head()
```

Melihat 5 baris terbawah di dataset, menggunakan fungsi tail()

```
df_can.tail()
```

Melihat daftar dari kepala kolom dataframe, gunakan .coloumns

```
df_can.columns.values
```

default dari index bukanlah list, maka kita ubah dahulu ke list

```
df_can.columns.tolist()
df_can.index.tolist()

print (type(df_can.columns.tolist()))
print (type(df_can.index.tolist()))
```

Untuk melihat dimensi dari dataframe, gunakan parameter .shape

```
# size of dataframe (rows, columns)
df_can.shape    
```

Untuk menghapus kolom yang tidak digunakan, bisa menggunakan method dari pandas, yaitu drop()

```
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

Merubah nama kolom, menggunakan method rename()

```
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

Lalu akan menambah kolom total 

```
df_can['Total'] = df_can.sum(axis=1)
```

Membuat ringkasan singkat dari setiap kolom dari dataframe
